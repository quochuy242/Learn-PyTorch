## Random tensors

Why is random tensors important?

Because the way many NN learn is that they start with tensors full of random number and then adjust those random number to better represent the data

```python
# Create a random tensor of size (3, 4)
random_tensor = torch.rand(3, 4)
random_tensor
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(3, 224, 224))
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

```python
# Permuting to image's channel from first dimension to third dimension
plt.imshow(random_image_size_tensor.permute(1, 2, 0))
```

## A range of tensor and tensors-like

```python
one_to_ten = torch.arange(0, 11, step=2)
one_to_ten
```

```python
ten_zeros = torch.zeros_like(input=one_to_ten)
ten_zeros
```
